```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt


ref_dat = genfromtxt("D:\Research_work\My_covid_research/approximation_model/future_data_ref.csv", delimiter=',')
# ref_dat=ref_dat[1:,1:]


proposed_dat = genfromtxt("D:\Research_work\My_covid_research/approximation_model/future_data_est.csv", delimiter=',')
# proposed_dat = genfromtxt("states_est.csv", delimiter=',')


tcn_dat = genfromtxt("D:\Research_work\My_covid_research/validation/forecasting/tcn_predicted_states.csv", delimiter=',')
tcn_dat=tcn_dat[1:,1:]


tft_dat=genfromtxt("D:\Research_work\My_covid_research/validation/forecasting/TFT_predicted_states.csv", delimiter=',')
tft_dat=tft_dat[1:,1:]


transformer_dat=genfromtxt("D:\Research_work\My_covid_research/validation/forecasting\Transformer_predicted_states.csv", delimiter=',')
transformer_dat=transformer_dat[1:,1:]

days=np.arange(1,len(ref_dat)+1)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(4, 4), dpi=150)
plt.plot(days, ref_dat[:,0], '.',color='black',label="$\mathrm{Ref}$")
plt.plot(days, tcn_dat[:,0], label="$M-1$")
plt.plot(days, transformer_dat[:,0], label="$M-5$")
plt.plot(days, tft_dat[:,0], label="$M-6$")
plt.plot(days, proposed_dat[:,0], label="$M-8$")
# plt.plot(days[0:extender], state_dat_est[0:extender,0], label="Suspectible_est")
plt.legend()
plt.grid()
plt.yticks(np.linspace(0.9997*np.min(ref_dat[:,0]), 1.000799*np.max(ref_dat[:,0]), 20))
plt.xticks(days)
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=6)
plt.title('Suscpectible')
```

```python
def smape(a, f):
    return 1/len(a) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))*100)

def mape(a,f):
    return np.mean(np.abs((a - f) / a)) * 100


print("SMPAE for Proposed Model: ", smape(ref_dat[:,0],proposed_dat[:,0]))
print("SMPAE for TFT Model: ", smape(ref_dat[:,0],tft_dat[:,0]))
print("SMPAE for TCN Model: ", smape(ref_dat[:,0],tcn_dat[:,0]))
print("SMPAE for Transformer Model: ", smape(ref_dat[:,0],transformer_dat[:,0]))
```
